```python


# Adjusting based on the actual column names
gb = df.groupby("ParentEduc").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore": "mean"})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot=True)
plt.show()
```

```python

gb1 = df.groupby("ParentMaritalStatus").agg({
    "MathScore": 'mean',
    "ReadingScore": "mean",
    "WritingScore": "mean"
})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1,annot=True)
plt.show()
```

from the above chart we have concluded that there is no/negligible impact on the student's score due to their parent's marital status

```python
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt

groupA = df[df['EthnicGroup'] == "groupA"].shape[0]
groupB = df[df['EthnicGroup'] == "groupB"].shape[0]
groupC = df[df['EthnicGroup'] == "groupC"].shape[0]
groupD = df[df['EthnicGroup'] == "groupD"].shape[0]
groupE = df[df['EthnicGroup'] == "groupE"].shape[0]

counts = [groupA, groupB, groupC, groupD, groupE]
labels = ["group A", "group B", "group C", "group D", "group E"]

plt.pie(counts, labels=labels, autopct="%1.2f%%")
plt.title("Distribution of Groups")
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the data
ax = sns.countplot(data=df, x='EthnicGroup')

# Adding labels to the bars
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                 (p.get_x() + p.get_width() / 2., p.get_height()), 
                 ha = 'center', va = 'center', 
                 xytext = (0, 10), 
                 textcoords = 'offset points')

plt.show()
```
